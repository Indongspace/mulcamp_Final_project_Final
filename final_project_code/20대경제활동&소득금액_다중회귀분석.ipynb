## 1. 라이브러리 및 데이터 확인

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/for_ML/연령별 경제활동인구 20대.csv')
df
```

## 2. 데이터 전처리

```python
# 20 / 30 데이터의 단위수가 천 명이기 때문에 1000 곱하기 함수
def multiply_1000(df, column_name):
    df[column_name] = df[column_name]*1000
    return df
```

```python
# 20 / 30 단위수 반영
df = multiply_1000(df, '경제활동인구')
df = multiply_1000(df, '취업자')
df = multiply_1000(df, '실업자')
df = multiply_1000(df, '비경제활동인구')
df.head()
```

```python
# 근로소득금액 데이터의 단위수가 백만이기 때문에 1000000 곱하기 함수
def multiply_million(df, column_name):
    df[column_name] = df[column_name]*1000000
    return df
```

```python
df = multiply_million(df, '근로소득금액')
df
```

## 3. 다중회귀분석

### 1) sklearn 라이브러리 활용

```python
from sklearn.linear_model import LinearRegression
```

```python
x = df.drop(columns=['근로소득금액'])
y = pd.DataFrame(df['근로소득금액'])
```

```python
x.head()
```

```python
y.head()
```

- 회귀식 : y = b0 + b1x1 + b2x2 + b3x3 + b4x4 + b5x5 + b6x6 + b7x7
- (x1 = 경제활동인구, x2 = 취업자, x3 = 실업자, x4 = 비경제활동인구, x5 = 경제활동참가율[%], x6 = 실업률[%], x7 = 고용률[%])

```python
# 모델링
model = LinearRegression()
model.fit(x,y)
```

```python
# 회귀분석 관련 지표 출력

# 1. Rsq(결정계수)
round(model.score(x,y),2)
```

```python
# 2. 회귀계수 출력
np.round(model.coef_,2)
```

```python
# 3. 회귀계수(절편)
np.round(model.intercept_, 2)
```

- 회귀식
- y = 8.08168445e+14 - 3.35955191e+08x + 5.24703178e+08 - 7.72477803e+07x - 1.42777777e+08x + 8.25552845e+12 + 2.16546184e+12x - 2.68789821e+13x

### 2) statsmodels 라이브러리 사용

```python
# statsmodel.fomula 활용
import statsmodels.api as sm
```

```python
# 모델링

x = sm.add_constant(x)
model = sm.OLS(y,x).fit()
summary = model.summary()
print(summary)
```

```python
- [1] 표준 오차는 오차의 공분산 행렬이 올바르게 지정되었다고 가정합니다.
표준 오차는 회귀 계수의 추정치의 신뢰성을 나타내는 중요한 지표입니다. 이 가정은 다음을 의미합니다:

회귀 모델에서 잔차(오차)들이 정규 분포를 따르고, 평균이 0이며 일정한 분산을 가져야 합니다.
잔차들 간에 자기상관(autocorrelation)이 없고, 독립적이어야 합니다.

해결 방법:

잔차의 정규성 검사: 잔차의 히스토그램과 Q-Q 플롯을 통해 잔차가 정규 분포를 따르는지 확인합니다.

이분산성(Heteroscedasticity) 검사: 잔차의 분산이 일정한지 확인하기 위해 Breusch-Pagan 테스트나 White 테스트를 수행합니다.

자기상관 검사: Durbin-Watson 통계량을 통해 잔차들 간의 자기상관을 확인합니다.

- [2] 가장 작은 고유값이 9.8e-28입니다. 이는 강한 다중공선성 문제가 있거나 설계 행렬이 특이하다는 것을 나타낼 수 있습니다.
고유값이 매우 작은 경우, 독립 변수들 간에 강한 다중공선성이 존재하거나, 설계 행렬이 거의 선형 종속적임을 나타냅니다. 이로 인해 회귀 계수의 추정치가 불안정해지고 해석이 어려워질 수 있습니다.

해결 방법:

다중공선성 확인: 분산 팽창 계수(VIF)를 계산하여 각 독립 변수의 다중공선성 정도를 평가합니다. VIF 값이 10을 넘으면 다중공선성이 높다고 판단합니다.

변수 선택: 다중공선성이 높은 변수를 제거하거나, 변수 선택 방법(예: 전진 선택법, 후진 제거법, 단계적 선택법)을 사용하여 모델을 단순화합니다.
주성분 분석(PCA): 고차원 데이터의 차원을 축소하여 다중공선성을 해결할 수 있습니다.

추가 설명

잔차의 정규성 검사: 시각적으로 확인하는 것이 중요합니다. 정규 분포를 따를 경우, 잔차의 히스토그램은 종 모양을 띠고, Q-Q 플롯에서는 점들이 대각선에 가깝게 위치합니다.

이분산성 검사: 이분산성이 존재할 경우, 회귀 계수의 표준 오차가 과소 혹은 과대 추정되어 신뢰구간이 잘못될 수 있습니다.

자기상관 검사: Durbin-Watson 통계량 값이 2에 가까울수록 잔차 간에 자기상관이 없음을 나타냅니다.
이를 통해 다중 회귀 분석의 가정이 만족되는지 확인하고, 다중공선성 문제를 해결하여 신뢰할 수 있는 회귀 모델을 구축할 수 있습니다.
```
